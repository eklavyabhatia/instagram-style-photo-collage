# Step-1
## Loading the images

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
top_left_img = cv2.imread('./images/top_left.jpg')
top_right_img = cv2.imread('./images/top_right.jpg')
bottom_left_img = cv2.imread('./images/bottom_left.jpg')
bottom_right_img = cv2.imread('./images/bottom_right.jpg')
center_img = cv2.imread('./images/center.jpeg')
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

rgb_top_left_img = convertToRGB(top_left_img)
rgb_top_right_img = convertToRGB(top_right_img)
rgb_bottom_left_img = convertToRGB(bottom_left_img)
rgb_bottom_right_img = convertToRGB(bottom_right_img)
rgb_center_img = convertToRGB(center_img)
```

```python
plt.imshow(rgb_top_left_img)
```

```python
plt.imshow(rgb_top_right_img)
```

```python
plt.imshow(rgb_bottom_left_img)
```

```python
plt.imshow(rgb_bottom_right_img)
```

```python
plt.imshow(rgb_center_img)
```

# Step - 2
## Resizing the images

```python
def resize(img,h,w):
    return cv2.resize(img,(h,w))

rgb_top_left_img = resize(rgb_top_left_img,200,200)
rgb_top_right_img = resize(rgb_top_right_img,200,200)
rgb_bottom_left_img = resize(rgb_bottom_left_img,200,200)
rgb_bottom_right_img = resize(rgb_bottom_right_img,200,200)
rgb_center_img = resize(rgb_center_img,100,100)
```

```python
plt.imshow(rgb_top_left_img)
plt.show()
plt.imshow(rgb_top_right_img)
plt.show()
plt.imshow(rgb_bottom_left_img)
plt.show()
plt.imshow(rgb_bottom_right_img)
plt.show()
plt.imshow(rgb_center_img)
plt.show()

```

# Step - 3
## Adding Borders

```python
side_border = np.full((200,10,3),0)
rgb_top_left_img = np.hstack((side_border,rgb_top_left_img))
```

```python
plt.imshow(rgb_top_left_img)
plt.show()
```

```python
rgb_bottom_left_img = np.hstack((side_border,rgb_bottom_left_img))
plt.imshow(rgb_bottom_left_img)
```

```python
rgb_top_right_img = np.hstack((side_border,rgb_top_right_img))
rgb_top_right_img = np.hstack((rgb_top_right_img,side_border))
plt.imshow(rgb_top_right_img)
plt.show()
```

```python
rgb_bottom_right_img = np.hstack((side_border,rgb_bottom_right_img))
rgb_bottom_right_img = np.hstack((rgb_bottom_right_img,side_border))
plt.imshow(rgb_bottom_right_img)
plt.show()
```

```python
hor_border = np.full((10,210,3),0)
rgb_top_left_img = np.vstack((hor_border,rgb_top_left_img))
plt.imshow(rgb_top_left_img)
plt.show()
```

```python
rgb_bottom_left_img = np.vstack((hor_border,rgb_bottom_left_img))
rgb_bottom_left_img = np.vstack((rgb_bottom_left_img,hor_border))
plt.imshow(rgb_bottom_left_img)
plt.show()
```

```python
hor_border = np.full((10,220,3),0)
rgb_top_right_img = np.vstack((hor_border,rgb_top_right_img))

rgb_bottom_right_img = np.vstack((hor_border,rgb_bottom_right_img))
rgb_bottom_right_img = np.vstack((rgb_bottom_right_img,hor_border))
plt.imshow(rgb_top_right_img)
plt.show()
plt.imshow(rgb_bottom_right_img)
plt.show()
```

# Step - 4
## Merging images

```python
final_top_img = np.hstack((rgb_top_left_img,rgb_top_right_img))
final_bottom_img = np.hstack((rgb_bottom_left_img,rgb_bottom_right_img))

plt.imshow(final_top_img)
plt.show()
plt.imshow(final_bottom_img)
plt.show()
```

```python
final_img = np.vstack((final_top_img,final_bottom_img))
plt.imshow(final_img)
plt.show()
```
